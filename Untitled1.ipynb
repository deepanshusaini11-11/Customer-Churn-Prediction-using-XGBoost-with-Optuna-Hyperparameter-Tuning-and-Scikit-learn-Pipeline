```python
import pandas as pd

def data_quality_check(df):
    quality_report = {
        'total_records': len(df),
        'duplicate_rows': df.duplicated().sum(),
        'missing_values': df.isnull().sum().to_dict(),

        # Invalid numeric checks
        'invalid_age': (df['Age'] < 0).sum(),
        'invalid_tenure': (df['Tenure'] < 0).sum(),
        'invalid_usage_frequency': (df['Usage Frequency'] < 0).sum(),
        'invalid_support_calls': (df['Support Calls'] < 0).sum(),
        'invalid_payment_delay': (df['Payment Delay'] < 0).sum(),
        'invalid_total_spend': pd.to_numeric(df['Total Spend'], errors='coerce').isnull().sum(),

        # Invalid categorical / empty checks
        'empty_gender': (df['Gender'].astype(str).str.strip() == '').sum(),
        'empty_subscription_type': (df['Subscription Type'].astype(str).str.strip() == '').sum(),
        'empty_contract_length': (df['Contract Length'].astype(str).str.strip() == '').sum(),

        # Date / interaction check
        'invalid_last_interaction': pd.to_datetime(df['Last Interaction'], errors='coerce').isnull().sum(),

        # Target column check
        'invalid_churn': df['Churn'].isnull().sum()
    }

    return quality_report
```

```python
quality_report = data_quality_check(df)
print(quality_report)
```
